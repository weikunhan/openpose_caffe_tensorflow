```python
# -*- coding: utf-8 -*-
""" Dump Caffe numpy layers

This module dumped information from Caffe model of numpy data structure. 

################################################################################
# Author: Weikun Han <weikunhan@gmail.com>
# Crate Date: 04/15/2018
# Update:
# Reference: 
################################################################################
"""

import numpy as np
import os

# Please modify input path to locate you file
CAFFE_DIR = 'caffe/'
LAYERS_OUTPUT = 'caffe/layers'

# Check location to save datasets
if not os.path.exists(LAYERS_OUTPUT):
    os.makedirs(LAYERS_OUTPUT)
    
# Setup input and output name
caffe_numpy_filename = 'openpose_pruned_50p.npy'

# Load Caffe network
caffe_numpy = os.path.join(CAFFE_DIR, caffe_numpy_filename)
caffe_data = np.load(caffe_numpy, encoding='bytes').item()

# Check for each layer name and each input data shapes 
for name, data in caffe_data.items():
    
    print('{:<5}: {}'.format(name, list(data.values())[0].shape))

print('------------------------Beginning dumping------------------------------')
    
# Write out weight matrices and bias vectors
for name, data in caffe_data.items():
    np.save(os.path.join(LAYERS_OUTPUT, 'w_{:s}.npy'.format(name)), 
            list(data.values())[0])
    np.save(os.path.join(LAYERS_OUTPUT, 'b_{:s}.npy'.format(name)), 
            list(data.values())[1])
    
    print('{:<5}:  weight-{} bias-{}'.format(name, 
                                             list(data.values())[0].shape, 
                                             list(data.values())[1].shape))

print('-------------------------Finished dumping------------------------------')
    
```
