```python
# -*- coding: utf-8 -*-
""" Dump Keras numpy layers

This module dumped information from "purned" Keras model. 

################################################################################
# Author: Weikun Han <weikunhan@gmail.com>
# Crate Date: 04/15/2018
# Update:
# Reference: 
################################################################################
"""

import numpy as np
import os

# Please modify input path to locate you file
KERAS_DIR = 'keras'
    
# Setup input and output name
keras_numpy_filename = 'openpose_pruned_50p.npy'

# Load Caffe network
keras_numpy = os.path.join(KERAS_DIR, keras_numpy_filename)
keras_data = np.load(keras_numpy, encoding='bytes').item()

# Check for each layer name and each input data shapes 
for name, data in keras_data.items():
    
    print('{:<5}: {}'.format(name, list(data.values())[0].shape))

print('-------------------------Finished dumping------------------------------')
```

Base on the above information, you need change you model and import it.

```python
from model_pruned_50p import openpose_model

# Setup input and output name
keras_model_filename = 'openpose_pruned_50p.h5'

# Get the skeleton of the Keras model
model = openpose_model()

# Write out weight matrices and bias vectors
for name, data in keras_data.items():
    
    # Look for weight and bias have same name between input data and model
    for layer in model.layers:
        if layer.name is name:
            w = list(data.values())[0]
            b = list(data.values())[1]
    
            # Store into Keras model
            layer_weights = [w, b]
            layer.set_weights(layer_weights)
            
# Output Keras model
keras_model = os.path.join(KERAS_DIR, keras_model_filename)
model.save(keras_model)

print('-------------------------Finished dumping------------------------------')
    
```
